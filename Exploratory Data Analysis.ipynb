## <p style="background-color: #2c3e50; ; color: white; padding-right: 100vw; text-align: center; font-family:newtimeroman;padding: 16px; border-radius: 15px 50px; border: 4px solid  black;font-size: 33px; text-shadow: 2px 2px 5px rgba(4, 5, 0, 0.5); box-shadow: 5px 5px 15px white;"><span style="color: #e67e22; margin: 0 10px; animation: flicker 1s infinite;">üçÇ</span> User Dataset üçÉ



```python
# List of columns to fill
columns_to_fill = ['Gender', 'Country', 'Degree', 'city', 'zip']

# Initialize SimpleImputer (you can choose strategy like 'most_frequent', 'constant', etc.)
imputer = SimpleImputer(strategy='most_frequent')  # Fill with most frequent value

# Apply imputer to specified columns
df1[columns_to_fill] = imputer.fit_transform(df1[columns_to_fill])

df1['isFromSocialMedia'].fillna(df1['isFromSocialMedia'].mode()[0], inplace=True)  # Mode
```

## <p style="background-color: #2c3e50; ; color: white; padding-right: 100vw; text-align: center; font-family:newtimeroman;padding: 16px; border-radius: 15px 50px; border: 4px solid  black;font-size: 33px; text-shadow: 2px 2px 5px rgba(4, 5, 0, 0.5); box-shadow: 5px 5px 15px white;"><span style="color: #e67e22; margin: 0 10px; animation: flicker 1s infinite;">üçÇ</span> Opportunity Wise Data üçÉ



```python
# Filling missing values: 0 for float columns, 'nill' for object columns
# Filling missing values based on data type
df.fillna({
    'Reward Amount': 0,
    'Badge Id': 'nill',
    'Badge Name': 'nill',
    'Skill Points Earned': 0,
    'Skills Earned': 'nill'
}, inplace=True)

# Convert to datetime format (optional, if dates are in string format)
df['Opportunity Start Date'] = pd.to_datetime(df['Opportunity Start Date'])

# Fill missing values using forward fill, then backward fill
df['Opportunity Start Date'].fillna(method='ffill', inplace=True)  # Forward fill



# Fill missing values using forward fill, then backward fill
df['Graduation Date(YYYY MM)'].fillna(method='ffill', inplace=True)  # Forward fill



# Filling missing values
df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)  # Mode
df['City'].fillna(df['City'].mode()[0], inplace=True)      # Mode
df['State'].fillna(df['State'].mode()[0], inplace=True)    # Mode
df['Zip Code'].fillna(0, inplace=True)                     # Fill with 0
df['Current Student Status'].fillna(df['Current Student Status'].mode()[0], inplace=True)  # Mode
df['Current/Intended Major'].fillna(df['Current/Intended Major'].mode()[0], inplace=True)  # Mode
# Convert dates to datetime format, with error handling
df['Apply Date'] = pd.to_datetime(df['Apply Date'], errors='coerce')
df['Opportunity Start Date'] = pd.to_datetime(df['Opportunity Start Date'], errors='coerce')
df['Graduation Date(YYYY MM)'] = pd.to_datetime(df['Graduation Date(YYYY MM)'], errors='coerce')
```
